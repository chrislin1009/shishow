```python
import pandas as pd

data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}

df = pd.DataFrame(data)

df
```

## 取得 Name 和 Salary 兩欄資料

```python
df[['Name', 'Salary']]
```

## 取出索引值 5 到 8 共四筆資料

```python
df.loc[5:8]
```

```python
df.iloc[5:9]
```

## 找出年齡和薪水都高於平均值的人員

```python
age_ave = df['Age'].mean()
salary_ave = df['Salary'].mean()
print("ageave: " , age_ave)
print("salaryave:" , salary_ave)
df[(df['Age'] > age_ave) & (df['Salary'] > salary_ave)]
```

## 公司人員的平均薪水、及最高、最低薪水、中位數

```python
avesalary = df['Salary'].mean()
maxsalary = df['Salary'].max()
minsalary = df['Salary'].min()
mediansalary = df['Salary'].median()

print("平均薪水: ", avesalary)
print("最高薪水: ", maxsalary)
print("最低薪水: ", minsalary)
print("中位數薪水: ", mediansalary)
```

## 公司共有幾個部門？

```python
departmentcount = df['Department'].unique()
print("部門數量: ", len(departmentcount))
```

## 哪個部門薪水最高？

```python
maxsalarydepartment = df.groupby('Department')['Salary'].mean().idxmax()
print("最高薪的部門: ",maxsalarydepartment,"部門")
```

## 哪個部門平均年齡最年輕

```python
minagedepartment = df.groupby('Department')['Age'].mean().idxmin()
print("最年輕的部門: ", minagedepartment , "部門")
```

## 年齡是否與薪資有關？

```python
corr_matrix = df[['Age', 'Salary']].corr()
if corr_matrix.iloc[0, 1] > 0.6:
    print("年齡與薪資有關")
else:
    print("年齡與薪資無關")
```
